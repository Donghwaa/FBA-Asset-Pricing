```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
brk['BRK-RF'] = brk['BRK.A']-brk['RF']
brk['MKT-RF'] = brk['MKT']-brk['RF']
```

2. DF Residuals : 잔차의 자유도  
전체 표본 수에서 측정되는 변수들(종속변수 및 독립변수)의 개수를 뺸것.    
cf. 자유도 : 조정된 표본, 표분분산의 평균이 모분산이 되도록 하기 위한 개념
>case) 180 - (3 + 1) = 176, 180개의 표본 중 측정되는 변수(제약조건 4개)를 제외 한 176개로 잔차의 분산을 구한다.  

3. DF Models : 독립변수의 개수

4. R-squared : 결정계수  
최소제곱법(MLS, Method of Least Squares)을 활용한 선형 회귀분석에서 종속변수의 분산(SST) 중에서 추정값(회귀값)의 분산(SSR)의 비율을 의미한다. -> 회귀 변동이 실제 변동(분산)을 얼마만큼 포착하는가를 의미한다. $$R^2 = \frac{SSR}{SST} = \frac{\textstyle\sum(\hat{y_i} - \overline{y})^2}{\textstyle\sum(y_i - \overline{y})^2}$$
>case) $R^2$ = 0.62, 실제 분산이 1이라고 했을 때 추정값(회귀값)의 분산은 0.62, 실제 변동의 62%를 포착했다(설명할 수 있다). 


5. Adj. R-squared : 수정 결정계수  
* 결정계수의 한계  
결정계수는 독립변수 개수가 많아질 수록 그 값이 커지게 된다. 종속변수의 변동을 설명해 주지 못하는 변수가 모형에 추가된다고 하더라도 값이 커질 수 있다.  
* 수정 결정계수  
다중회귀분석(독립변수가 2개이상)을 수행하는 경우 결정계수의 한계를 보완하기 위해 표본의 크기(n)와 독립변수의 수(p)를 고려하여 수정된 결정계수를 구한다. $$Adj.R^2 = 1 - \frac{(n-1)}{(n-p-1)}(1-R^2)$$
>case)$$Adj.R^2 = 1 - \frac{(180-1)}{(180-3-1)}(1- 0.62) = 0.6135$$  


6. F-statistic & Prob (F-statistic) : 회귀모델 전체의 통계적 유의성 검정
$$y = \beta_0 + \beta_1x_0 + \beta_2x_1 +\epsilon$$

$$\begin{cases}
   H_0 : \beta_0 = \beta_1 = \beta_2 = 0 \\
   H_1 : 모든 \beta_i (i = 0,1,2) 가 \space 0은 아니다.
\end{cases}$$
>Case : Prob (F-statistic) 7.88e-37 < 0.05
>1. Decision : $H_0$ 기각한다
>2. Conclusion : 모델의 회귀계수 중 적어도 하나는 0이 아니다 -> 모델 전체의 회귀성 존재한다.

7. AIC & BIC
Data를 이용하여 모형만들 때, Overfit된 모형(복잡한 모형)을 만든다면, 새로운 Data에 대해 설명력이 떨어진다. 따라서 복잡한 모형에 대한 패널티를 부여하여 좀 더 Roubust한 모형을 선택하게 하는 기준이 되는 지표다.  더 작은 값을 가진 모형을 선택한다
* AIC : 실제 Data와 모형을 통해 예측한 Data 분포가 동일한지 판단 $$AIC = -2log(likelihood) + 2p$$
* BIC : 실제 Data 표본의 크기도 고려하여 분포의 동일성 판단 (표본이 큰 경우 사용) $$BIC = -2log(likelihood) + plog(n)$$
where, p = 변수 개수, n : 데이터 수
>Case : Constant 항이 있을 땐 변수에 + 1을 해준다
>* AIC = -2*(391.72) + 2*(3+1) = -775.44
>* BIC = -2*(391.72) + (3+1)*log(180) = -762.66
