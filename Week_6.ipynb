```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
brk = pd.read_csv('BRK13F.csv', header = 0, index_col = 0)/100
brk.index = pd.to_datetime(brk.index).to_period('M')
```

```python
brk
```

```python
brk['BRK-RF'] = brk['BRK.A']-brk['RF']
brk['MKT-RF'] = brk['MKT']-brk['RF']
```

```python
brk
```

```python
Y1 = brk['BRK-RF']
X1 = brk[['MKT-RF', 'SMB', 'HML']]
X1 = sm.add_constant(X1, has_constant = 'add')
model = sm.OLS(Y1,X1).fit()
print(model.summary())
```

```python
Y2 = brk['BRK-RF']
X2 = brk[['MKT-RF', 'SMB', 'HML','UMD']]
X2 = sm.add_constant(X2, has_constant = 'add')
model = sm.OLS(Y2,X2).fit()
print(model.summary())
```

```python
Y3 = brk['BRK-RF']
X3 = brk[['MKT-RF', 'SMB', 'HML','QMJ','BAB','UMD']]
X3 = sm.add_constant(X3, has_constant = 'add')
model = sm.OLS(Y3,X3).fit()
print(model.summary())
```
